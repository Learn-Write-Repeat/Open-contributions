### Step 1- Importing the required libraries

```python
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

### Step-2 Loading the dataset
>This is a preprocessed dataset which was downloaded from a third party website where they had obtained all the distinct words from a NLP dataset and put them into individual columns and each row represents the emails. The columns are filled with the count of each word occuring in an email. 

```python
df = pd.read_csv(r"C:\Users\Nikhitha Siva\Desktop\emails.csv")
df.head(10)
```

### Step-3 Checking for missing values

```python
df.isnull().sum() 
```

### Step-4 Insights about the data

```python
df.describe()
```

```python
df.corr()
```

### Step-5 Assigning the independent variables to X and dependent variables to Y. 

```python
X = df.iloc[:,1:3001]
X
```

```python
Y = df.iloc[:,-1].values
Y
```

### Step-6 Splitting the data into train and test

```python
train_x,test_x,train_y,test_y = train_test_split(X,Y,test_size = 0.25)
```

### Step-7 Building the model and predicting

```python
mnb = MultinomialNB(alpha=1.9)  #calling the model       
mnb.fit(train_x,train_y)        #training the model using the training data
y_pred1 = mnb.predict(test_x)   #predicting the result using the independent variables.
accuracy_score(y_pred1,test_y)  #obtaining the accuracy of the model by comparing the predicted results and test dataset
```

>The accuracy is 94%.That's a pretty good model that we have built.
Now we have succesfully built a model to classify whether an email is spam or not!
