linspace :
y = linspace(a,b,n) generates a row vector y of n points linearly spaced between and including a and b.

We can plot many graphs in a single graph:

```python
x = np.linspace(3, 7, 500)

fig, ax = plt.subplots()  
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic') 
ax.plot(x, x**3, label='cubic') 
ax.set_xlabel('x label')  #Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Plot")  # Add a title to the axes.
ax.legend()  
```

```python
x = np.linspace(1, 2, 5)

fig, ax = plt.subplots()  
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic') 
ax.plot(x, x**3, label='cubic') 
ax.set_xlabel('x label')  #Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Plot")  # Add a title to the axes.
ax.legend()  
```

```python
x = [1,2,3] 
y = [2,4,6] 

plt.plot(x, y, label = "line1") 

x1 = [1,2,3] 
y1 = [4,8,3] 

plt.plot(x1, y1, label = "line2") 

plt.xlabel('x-axis') 

plt.ylabel('y-axis') 

plt.title('Two lines on one graph') 

plt.legend() 

plt.show() 
```

<b>Pie-chart</b>

```python
company = ['DevIncept', 'Adobe', 'Infosys', 'Amazon'] 

s = [8, 3, 4, 6] 

colors = ['r', 'y', 'g', 'b'] 

plt.pie(s, labels = company, colors=colors, startangle=90, shadow = True, explode = (0.2, 0, 0, 0), radius = 1.2, autopct = '%1.1f%%') 

plt.legend() 

plt.show() 
```

<b>Contour graph</b>

```python
xlist = np.linspace(-10.0, 10.0, 100)
ylist = np.linspace(-5.0,53.0, 100)
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2 + Y**2)
fig,ax=plt.subplots(1,1)
cp = ax.contourf(X, Y, Z)
fig.colorbar(cp) # Add a colorbar to a plot
ax.set_title('Filled Contours Plot')
#ax.set_xlabel('x (cm)')
ax.set_ylabel('y (cm)')
plt.show()
```

<b>Box Plot</b>

```python
np.random.seed(10) 
data = np.random.normal(50, 20, 100) 
plt.boxplot(data) 
plt.show() 
```

```python
np.random.seed(10) 
data_1 = np.random.normal(10, 10, 500) 
data_2 = np.random.normal(80, 20, 500) 
data_3 = np.random.normal(40, 30, 500) 
data_4 = np.random.normal(70, 40, 500) 
data = [data_1, data_2, data_3, data_4] 

fg = plt.figure(figsize =(10, 7)) 

axis = fg.add_axes([0, 0, 1, 1]) 
 
bp = axis.boxplot(data) 

plt.show() 
```
