```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Function for Harris Corner Detection Method

```python
def harrisCorner(img, color=[0, 0, 0]):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray = np.float32(gray)  # harris function requires float32 value 
    
    detect = cv2.cornerHarris(gray, 5, 3, 0.04)
    detect = cv2.dilate(detect, None)
    img[detect > 0.01 * detect.max()] = color
    
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.title('Harris Corner Output')
    plt.show()
```

## Function for Shi Tomasi Corner Detection Method

```python
def shi_tomasi(img, color=[0, 0, 0], max_corners=23, min_dist=15):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    corners = cv2.goodFeaturesToTrack(gray, max_corners, 0.01, min_dist)
    corners = np.int64(corners)  # circle function requires int values and corners gets float values
    
    for i in corners:
        x, y = i[0][0], i[0][1]
        cv2.circle(img, (x, y), 5, color, -1)
        
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.title('Shi-Tomasi Output')
    plt.show()
```
