#                                    Machine Learning

### Introduction

   Machine learning is quite interesting field to study.It is all around us in this modern world. From Suggestions on Social media platforms to Google Maps for navigation,machine learning finds its applications in almost everywhere.
  
   We may think that Machine learning is a new technology/topic to adopt or discuss. But No,the concept of Machine learning came into picture in the year 1950, when Alan Turing published a paper answering the question "Can machines think?".

### What is Machine learning?
 Well,in simplest way, Machine learning is teaching the machine about something in a productive way.
 Machine learning(ML) is the field of computer science which makes the machine capable of learning on its own without being explicitly programmed.ML algorithms can learn on its own from past experiences,just like humans do.When exposed to new data,these algorithms learn,change and grow by themselves without you needing to change the code every single time.

### How do ML Algorithm works?
 ML algorithms utilize a variety of techniques to handle large amount of complex data to make decisions.These algorithms complete the task of learning from data with certain inputs given to the machine.It all starts with cleaning the collected data,training and testing the data to get patterns.After this,we give new input data to make predictions.And then we evaluate the accuracy of predictions and results. 

### Classification of ML Algorithms and its Applications

Machine Learning algorithms are classified according to their purpose and how a program learns to become more accurate and precise in its predictions.And its main categories are the following:                   
 1.Supervised learning                                                                                                         
 2.Unsupervised learning                                                                                                       
 3.Reinforcement learning 

### Supervised Learning
#### *The outputs are labeled and the inputs are mapped to corresponding outputs*
Supervised machine learning algorithms are designed to learn by example under our supervision.It tries to figure out the relationship between input features and output targets.Based on the relationship, it predicts the output of new data.If the mapping is correct then the machine has learned successfully.If not,we need to make necessary corrections to the algorithms to learn it correctly.This algorithm is used for labelled dataset which has both input and output data(values).In this,we divide the data into train and test sets from which we train the model and test the data for predicting output and to see the accuracy.
                                                                                                                               
#### Types
 1.Classification                                                                                                              
 2.Regression
 
 #### 1.Classification
   Classification algorithms are all about to take an input value and assign it to a class or category that it fits into based on training data used.
   The most common example of classification is determining if an email is spam or not.In this,the algorithm is trained with data having emails both spam and not spam.The model will find the features within the data that correlates to either class and create the mapping function.Based on this function,the model will predict the new email whether it is spam or not spam.
   Depending on the situation and data, below are the few types of classification algorithms:                                      * Linear Classifiers
   * Support Vector Machines
   * Decision Trees
   * Nearest Neighbour
   * Random Forest                                                                                                                * Naive bayes                                                                                                                           
#### 2.Regression
    Regression algorithms tries to find the relationship between dependent and indepenent variables.Regression analysis is widely used for prediction and forecasting.There are various types of regressions which we use in data science and machine learning.Each type has its own importance on different scenarios.Few of them are:
     * Linear regression
     * Logistic regression
     * Polynomial regression
     * Random forest regression
     * Support vector regression
 

### Unsupervised Learning
  Unsupervised learning is technique where we do not need to supervise the model, unlike Supervised learning.Instead we have to allow the model to work on its own to discover information.It deals with unlabelled data.It finds all kinds of unknown patterns in data.
  For example,let us take the case of a baby and her pet or family dog.She knows and identifies this dog.A few months later,her family friend brings a dog along with him and tries to play with the baby.Baby has not seen this dog earlier but she recognizes many features like four legs,two ears etc.,are like her pet dog.She identifies a new animal which looks like a Dog at her home.This is Unsupervised Learning,where you are not taught but learned from the data.
#### Types of Unsupervised Learning
  * Clustering
  * Association
  
#### Clustering
 Clustering algorithm will process the data and find natural clusters(groups) if they exist in data.It is simply,grouping the data points having similar features.
#### Few Types of clustering
   * K-Means clustering
   * K-NN(K nearest neighbour)
   * Principal component analysis
#### Association
 Association rules allows us to establish associations amongst data objects inside large databases.It is about discovering interesting relationships between variables in large databases.
 Eg: Movie group by the rating given by the viewers.

### Reinforcement Learning
  Reinforcement learning is all about making decisions sequentially.In simple words we can say that the output depends on the state of current input and the next input depends on the output of the previous input.In reinforcement learning,decision is dependent.So we give labels to sequences of dependent decisions.
  eg: Chess game
#### Types of Reinforcement learning
   * Positive
   * Negative
