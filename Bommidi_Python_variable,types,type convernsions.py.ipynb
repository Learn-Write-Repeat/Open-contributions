# Set

```python
a = {1,2,3,4,4,4,3}
print(a)            # repeted elements are deleted
print(type(a))
```

```python
s = {"Dev" , "Incept" , 10 , 5.4}            # output order is not predicted in sets
s
```

```python
# Removing one element in the set

s = {10 , 'True' , 4.9 , 50}
s.remove(4.9)     
print(s)
```

```python
# An error will occur if we try to remove an element which is not there in the set

s = {10 , 'True' , 4.9 , 50} 
s.remove(101)
s
```

```python
# Using "discard" method, there will be no error even though that element is not there in the set 

s = {10 , 'True' , 4.9 , 50}
s.discard(101)
print("The length of the set is" ,len(s))
print(s)
```

```python
s = {10 , 'True' , 4.9 , 50}
s.clear()               # "clear" method deletes all the elements in the set
print(len(s))
```

# Operation on different types

***Arithemetic Operators***

```python
x = 10 
y = 20

print("Addition : ",x + y)
print("Subtraction : ",x - y)
print("Multiplication : ",x * y) 
print("Divivsion : ",x / y)
print("Modulus : ",x % y)
print("Floor division : ",x // y)
print("Exponential : ",x ** y)
```

***Comparison Operators***

```python
x = 10 
y = 20 

print("x = y :", x == y)
print("x < y :", x < y)
print("x > y :", x > y)
print("x <= y :", x <= y)
print("x >= y :", x >= y)
print("x != y :", x != y)     # not equal to

```

***Logical Operators***

```python
x = 10

print("AND :",x > 5 and x < 10)        # Both conditions True --> output will be True , or else False
print("OR :",x > 5 or x < 10)         # Any one condition is True --> output will be True , or else False
print("NAND :",not(x >5 and x < 10))    # opposite to AND i.e, NAND 
print("NOR : ",not(x > 5 or x < 10) )   # opposite to OR i.e, NOR
```

***Bitwise Operators***

```python
# These operators are used to compare  Binary numbers
x = 1001
y = 1000

print("x AND y is :",x & y)
print("x OR y is :",x | y)
print("x XOR y is :",x ^ y)
print("NOT y is :", ~ y)
```
