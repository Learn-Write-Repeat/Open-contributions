```python
import cv2
from matplotlib import pyplot as plt
```

```python
image = cv2.imread("test.jpg")
Image = cv2.cv2.cvtColor(image, cv2.COLOR_BGR2RGB)# As opencv has images in BGR but we want them in RGB in order to display using matplotlib.
#Loading the image
```

```python
cv2.imshow('image',image)#This line displays the image using opencv
plt.imshow(Image)#Displays the image in jupyter using matplotlib.
cv2.waitKey(0) #This line waits for us to press a key on the keyboard. Till then the program waits and it doesn't move forward.
cv2.destroyAllWindows()#Once we press any key, the program destroys all the windows.
```

```python
#As you can see by executing the above cells , we have loaded an image and we have simply displayed it.
```

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
```

```python
# Now we have loaded a cascade classifer in order to predict a face in the image.
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)#Converting the image to grayscale
```

```python
faces = face_cascade.detectMultiScale(gray, 1.01, 4,minSize = (10,10))#This line of code detects the faces, and returns the boundaries in the form of a list.
```

```python
# The parameters used  are the image, then a scale factor  that decides how much the size of the image reduces at each scale, then a min neighbours parameter and finnaly a min size. 
```

```python
# Now we display our results.
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 4)#Creates a bounding box around the face.
cv2.imshow("Image",image)#Press a key once a window pops up.
Image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) #Again as we know opencv uses BGR format and we need to convert to RGB for using imshow by matplotlib.
plt.imshow(Image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# We created a bounding box around our detected face and as you run this block you can see the face has been perfectly detected by the classifer.
```

```python
eye_cascade = cv2.CascadeClassifier("haarcascade_eye.xml")
img = cv2.imread("test2.jpg")
```

```python
#Loaded a cascade classifier in order to find an eye in the image
```

```python
Gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
eyes = eye_cascade.detectMultiScale(Gray)
```

```python
for (ex,ey,ew,eh) in eyes:
    cv2.rectangle(img,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)#Creates  bounding box around the eyes.
cv2.imshow("Image",img)#Press any key in order to close the pop up window.Displays image using  opencv's imshow.
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)#Converting to RGB for displaying using matplotlib.
plt.imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()  
```

```python
#As you can see this time we have shown eyes in an image. Thus you can always use cascade classifers in order to classify stuff in images.
```
